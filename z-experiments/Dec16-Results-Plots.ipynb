```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['BL', 'AUG']
random = [2.8, 2.5]
brown = [4.4, 4.7]
mattson = [4.6, 5.5]

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, random, width, label='Random', color='lightgrey')
rects2 = ax.bar(x, brown, width, label='Hand-Crafted', color='darkgrey')
rects3 = ax.bar(x + width, mattson, width, label='Ours')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Average Behaviors Discovered')
ax.set_ylim(0, 8)
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)
fig.set_figwidth(5)
fig.set_figheight(5)


plt.show()
```
